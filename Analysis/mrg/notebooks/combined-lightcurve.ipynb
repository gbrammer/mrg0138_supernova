```python
def show_lens_model_time_delays(ax, mkey):
    yl = ax.get_ylim()
    
    #\textit{C} & 6.324 & 9.526  & 6.879  & 180.3 & 82.7 &7402.8 &$-15.18\pm0.33$ \\
    #\textit{D} & 6.281 & 9.923  & 7.472  & 184.4 & 85.3 &7034.9 &$-13.46\pm0.31$ \\
    #\textit{E} & 8.942 & 18.48  & 13.257 & 103.6 & 50.2 &5876.8 &$-8.91\pm0.27$ \\
    #\textit{F} & 8.699 & 18.069 & 12.719 & 108.3 & 53.7 &6020.5 &$-9.67\pm0.28$ \\

    tdelay = {}
    #                  μ1     μ2      μ3   t2−t1  t3−t1 t4−t1, bayes
    tdelay['C'] = [6.324, 9.526 , 6.879 , 180.3 , 82.7, 7402.8 , -15.18]
    tdelay['D'] = [6.281, 9.923 , 7.472 , 184.4 , 85.3, 7034.9 , -13.46]
    tdelay['E'] = [8.942, 18.48 , 13.257, 103.6 , 50.2, 5876.8 , -8.91]
    tdelay['F'] = [8.699, 18.069, 12.719, 108.3 , 53.7, 6020.5 , -9.67]
    
    z = 1.9456
    t2 = im2['x'][0]
    
    t2_lo = im2['error_x']['arrayminus'][1]
    t2_hi = im2['error_x']['array'][1]

    #k = 'E'
    td = tdelay[mkey]
    dt1 = t2 + td[3]
    dt3 = dt1 - td[4]

    dy = (yl[1]-yl[0])
    
    ax.vlines(dt1, yl[0]+0.05*dy, yl[0]+0.2*dy, color=colors[0], alpha=0.8)
    ax.fill_between([dt1-t2_lo, dt1+t2_hi], yl[0]+0.05*dy, yl[0]+0.2*dy, color=colors[0], alpha=0.5)
    
    ax.vlines(dt3, yl[0]+0.05*dy, yl[0]+0.2*dy, color=colors[2], alpha=0.8)
    ax.fill_between([dt3-t2_lo, dt3+t2_hi], yl[0]+0.05*dy, yl[0]+0.2*dy, color=colors[2], alpha=0.5)
    
    ax.text(10, yl[0]+0.12*dy, r'Lens model $xxx$'.replace('xxx', mkey), fontsize=6, ha='right', va='center')
    
    print('Time delay ', mkey, t2, dt1, dt3)
    
```
